```python
num = 1  # counter variable intelize with 10

while num< 15: # coundition using the counter variable
    print("loop executing time: ", num-5)
    num = num + 1 # increment in the counter variable
```

```python
counter = 0
while counter < 5:
    print("loop count", counter)
    counter = counter + 1
```

```python
doc = 110
while doc > 60:
    print("loop count", doc)
    doc = doc - 5
```

```python
num = 10
while num >= 0:
    print("loop count", num)
    num = num - 2
```

```python
count = 5

while count > 0:
    print(count)
    count -= 1

print("Time's up!")
```
